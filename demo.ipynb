# Testing differentiability using taylor test 

```python
def f(x):
    return np.sin(x)
def df(x):
    return np.cos(x)
def ddf(x):
    return -np.sin(x)

for i in range(3):
    taylor_test(f,df,ddf, 1, float(1), order =i)
```

```python
def Rosenbock(x):
    return 100*(x[1]-x[0]**2)**2+(1-x[0])**2
def dRosenbock(x):
    return np.array([2*(200*x[0]**3-200*x[0]*x[1]+x[0]-1), 200*(x[1]-x[0]**2)])
def ddRosenbock(x):
    return np.array([[-400*(x[1]-x[0]**2)+800*x[0]**2+2, -400*x[0]], [-400*x[0], 200]])

taylor_test(Rosenbock, dRosenbock, ddRosenbock, [-1,1],[1,1], order=2)
```

# Example of a bad convergence of the gradient descent

Example 7.5 in [1]

```python
from Gradient_Descent import unconstrained_opt

def Rosenbock(x):
    return 100*(x[1]-x[0]**2)**2+(1-x[0])**2
def dRosenbock(x):
    return np.array([2*(200*x[0]**3-200*x[0]*x[1]+x[0]-1), 200*(x[1]-x[0]**2)])

solver=unconstrained_opt(Rosenbock,dRosenbock, maxiter=100000)

solver.solve(np.array([-1.2, 1]))
```

Gradient converges to minimum [1,1] after 19227 iterations.

## Gradient descent with Powell Wolfe stepsize 

Algorithm 9.3 in [1]

```python
solver=unconstrained_opt(Rosenbock,dRosenbock, step_size_choice="powell", maxiter=100000)
solver.solve(np.array([-1.2,1]))
```

Convergence after 18713 iterations

# Much better convergence with Newton method

```python
solver=unconstrained_opt(Rosenbock,dRosenbock, ddRosenbock, optimizer="Newton")
solver.solve(np.array([-1.2,1]))
```

BFGS (Quasi newtonmethod) fast convergence, without second derivative

```python
solver=unconstrained_opt(Rosenbock,dRosenbock, optimizer="BFGS", maxiter=2000)
solver.solve(np.array([-1.2,1]))
```
